```python
# Eksik değerleri nasıl doldurabiliriz?
"""Eksik değerleri ortalamayla doldurabiliriz. Ancak bu hangileri için anlamlı hangileri 
için anlamsız diye düşünürsek
veriyi incelediğimde bana göre ortalamayla doldurulması anlamlı olan yalnızca 4 kolon var. 
AVG_MATH_4_SCORE, AVG_MATH_8_SCORE, AVG_READING_4_SCORE, 
AVG_READING_8_SCORE. Bu değerler zaten ortalama değerler ve birbirlerine çok yakın. 
Dolayısıyla bu kolonlar boş kalmaması adına
rahatlıkla ortalamayla doldurulabilir.

Ancak diğer kolonlardaki ardışık verilerde tutarlı bir ilişki yok veya 
aşırı miktarda farklılık gösteriyorlar. Burada 
null değerleri ortalamayla değiştirmenin bir zararı olmayabilir 
ancak faydasının da olacağı söylenemez. """
```

```python
df.head()
```

```python
# Burada 1992 yılının TOTAL_REVENUE kolonundaki boşlukları ortalama değerle dolduracağız.
# Ortalama değeri şu şekilde bulalım.

df_1992 = df.loc[lambda data: df['YEAR'] == 1992]
mean_of_1992 = df_1992.TOTAL_REVENUE.mean()

# İkinci bir yöntem
mean_of_1992 = np.nanmean(df_1992.TOTAL_REVENUE)
print(mean_of_1992)
```

```python
# Burada 1992 yılının Nan değerleri aynı yılın ortalamasıyla değiştirilmiştir.
df_1992['TOTAL_REVENUE'] = df_1992['TOTAL_REVENUE'].fillna(mean_of_1992)
print(df_1992['TOTAL_REVENUE'])
print('Ortalama:', df_1992['TOTAL_REVENUE'].mean())
```

```python
# Basit bir fonksiyon ile lineer interpolasyon ile doldurma işlemi aşağıdaki gibi uygulanabilir. 
df_92 = df.loc[lambda data: df['YEAR'] == 1992]
df_92['TOTAL_REVENUE'] = df_92['TOTAL_REVENUE'].interpolate(limit=5, limit_direction='forward')
```

```python
inter_mean_92 = df_92.TOTAL_REVENUE.mean()
print('İnterpolasyon sonucundaki ortalama:', inter_mean_92)
print('Diğer ortalama: ', mean_of_1992)
```
