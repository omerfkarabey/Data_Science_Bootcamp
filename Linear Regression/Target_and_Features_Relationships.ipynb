### 2. Hava durumu modeli

```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('weatherHistory.csv')
df.head()
```

Hedef değişkeni oluşturalım.

```python
temp = df['Temperature (C)'] - df['Apparent Temperature (C)']
```

Model: 

temp = constant + beta1 * temperature(C) + beta2 * apparent_temperature(C) + Error

OLS uygulayarak modelimizin parametrelerini tahmin edelim.

```python
X = df[['Humidity','Wind Speed (km/h)']]
X
Y = temp

X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
results.summary()
```

OLS kullanarak modelimizi tahmin ettik. 

* Nem ve rüzgar hızı özelliklerinin hedef üzerinde etkisinin istatistiksel olarak anlamlı olduğunu görüyoruz.
* Özelliklerin katsayıları rakamsal olarak bilsek de hedefe etkisini yorumlamak, yalnızca bu veriyle güç.
* Bu iki değişken varyansın 0.238'lik bir kısmını açıklıyor. Bu modelimizin zayıf olduğu konusunda bir fikir verebilir. Ancak bu açıklanan varyans azlığı tahminlerimizin yanlı olup olmadığı hakkında kesin bir bilgi vermez. Bunu daha iyi anlamak için residual plot'lara bakılabilir. (ileride uygulanacaktır)

Modelimize nem ve rüzgar hızı etkileşimini de dahil edelim.

```python
hum_wind = df['Humidity'] * df['Wind Speed (km/h)']
df['hum_wind'] = hum_wind

X2 = df[['Humidity', 'Wind Speed (km/h)', 'hum_wind']]
Y2 = temp

X2 = sm.add_constant(X2)

results2 = sm.OLS(Y2,X2).fit()
results2.summary()
```

* Etkileşimli modelimizde tüm özellikler istatistiksel olarak anlamlı kalmaya devam etti. Ek olarak R-squared değerinin artması da açıklanan varyans miktarının arttığını gösteriyor. 

* Etkileşimli özelliği dahil ettikten sonra, etkileşimsiz ayrı ayrı hallerinin katsayılarının negatife döndüğünü gördük.

* İlk modelimizden nem ve rüzgar hızının hedef değişkenin değerini arttırdığını biliyoruz. Yani nem ve rüzgar hızı hissedilen sıcaklığı azaltıyor. 

* İkinci modelde nem ve rüzgar hızı katsayılarının işaret değiştirmesinin sebebi büyük ihtimalle çoklu doğrusallıktan kaynaklanıyor. Etkileşimli değişken diğer değişkenlerle de bir ilişkiye sahip. Böyle değişkenleri 'confounded' veya 'omitted' değişkenler olarak adlandırıyoruz. 

### 3. Ev fiyatları modeli

```python
hdf = pd.read_csv('train.csv')
hdf.head()
```

```python
hdf = hdf.drop(['MiscFeature', 'Alley', 'PoolQC', 'Fence','LotFrontage', 'FireplaceQu' ], axis = 1)
```

```python
hdf = hdf.dropna()
```

```python
X = hdf[['Street', 'Utilities', 'YearBuilt','LotArea', 'PoolArea', 'YrSold']]
hdf.Street = hdf.Street.replace({'Pave':1, "Grvl":0})
hdf.Utilities = hdf.Utilities.replace({'AllPub':1, 'NoSeWa':0})
```

Verideki değişkenleri standartlaştıralım.

```python
x = X[['YearBuilt', 'LotArea', 'PoolArea', 'YrSold']]
x_mean = x.mean()
x_std = x.std()
X_st = (x-x_mean)/x_std
X = X_st
X.head()
```

```python
Y = (hdf.SalePrice - hdf.SalePrice.mean())/hdf.SalePrice.std()
```

OLS kullanarak modelimizi tahmin edelim.

```python
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()
results.summary()
```

İstatistiksel anlamları P-value üzerinden incelediğimizde 'YrSold' yani satış yılının anlamsız olduğunu görebiliriz. İkinci modelimizde bunu çıkarıp parametrelerdeki değişimleri inceleyelim.

```python
X3 = X.drop(['YrSold', 'const'], axis = 1)
X3.head()
```

```python
X3 = sm.add_constant(X3)
results2 = sm.OLS(Y,X3).fit()
results2.summary()
```

* İkinci modelimizde satış yılını çıkardık. Tahminimiz üzerinde olumsuz bir etkisi olmadığını gördük.
* Özellik değişkenlerimiz hedef üzerinde pozitif bir etkiye sahip. 'YearBuilt' katsayısı en fazla olandır. Diğer değişkenlerden daha fazla bir etkiye sahip olduğunu düşünebiliriz.
* Satış yılının satış fiyatına doğrudan etkisi olması zaten mantıksızdı. Örneğin çok eski bir binanın yalnızca bu sene satıldığı için daha pahalı olmasını bekleyemeyiz.
* Havuz alanı açıkçası fazla etkili değil. Bu da normal diyebiliriz. 
* En çok etkili olan değişkenimiz 'YearBuilt', yapım yılı. Yeni evlerin eski evlerden pahalı olmalarından daha doğal bir durum yoktur.
